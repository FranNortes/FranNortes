## Analisis EDA

**1. Tendencia del Pop Koreano a lo largo de los tiempos (desde 1992-2020)**

En este gr√°fico de lineas observamos un crecimiento sostenido de produccion de videos musicales en el KPOP desde sus comienzos en el 1992 hasta 2020 con el pico mas alto de produccion en 2019, en la grafica podemos observar la produccion de videos musicales debido al Covid.

```python
Kpop_music_videos['Year'] = pd.to_datetime(Kpop_music_videos['Date']).dt.year
trend_df = Kpop_music_videos.groupby('Year').size().reset_index(name='Count')

plt.figure(figsize=(12, 6))
sns.lineplot(data=trend_df, x='Year', y='Count', marker='o')
plt.title('Evoluci√≥n del K-POP a lo largo del tiempo (1992-2020)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Videos', fontsize=12)
plt.grid(True)
plt.savefig('kpop_t')
```

**2. √âxito por g√©nero (grupos masculinos vs grupos femeninos)**

En este gr√°fico se puede observar la cantidad de grupos musicales activos haciendo una comparativa de grups musicales masculinos y femeninos. </br>
Observarmos que hay ligeramnete una mayor  cantidad de grupos masculinos que grupos femeninos.


```python
boy_groups_active = Kpop_idols_boys[Kpop_idols_boys['Active'] == 'Yes']
girl_groups_active = Kpop_idols_girls[Kpop_idols_girls['Active'] == 'Yes']

success_gender = pd.DataFrame({
    'Group Type': ['Boy Groups', 'Girl Groups'],
    'Active Groups': [len(boy_groups_active), len(girl_groups_active)]
})

plt.figure(figsize=(8, 6))
sns.barplot(data=success_gender, x='Group Type', y='Active Groups', palette='coolwarm')
plt.title('Cantidad de grupos activos seg√∫n el g√©nero', fontsize=14)
plt.xlabel('Group Type', fontsize=12)
plt.ylabel('Number of Groups', fontsize=12)
plt.savefig('success_by_gender.png')
plt.show()
```

**3. Compa√±√≠as l√≠deres en la producci√≥n de grupos**

En esta tercera grafica, nos ense√±a las 10 principales compa√±ias productoras de grupos Kpop segun el numero de grupos formados.</br> </br>
A la cabeza de esta grafica conatmos con la productora musical SM SM Entertainment con 12 grupos. </br>
Continuada de JYP Entertainment con 10 grupos y FNC en 3 lugar con 8 grupos. </br>
Esta grafica tambien nos muestra una igualacion de 5 grupos musicales en 5 compa√±ias.

```python
all_groups_df = pd.concat([Kpop_idols_boys, Kpop_idols_girls], ignore_index=True)
top_companies = all_groups_df['Company'].value_counts().head(10).reset_index()
top_companies.columns = ['Company', 'Group Count']

plt.figure(figsize=(10, 6))
sns.barplot(data=top_companies, x='Group Count', y='Company', palette='viridis')
plt.title('Principales compa√±√≠as productoras de grupos de K-POP', fontsize=14)
plt.xlabel('Number of Groups', fontsize=12)
plt.ylabel('Company', fontsize=12)
plt.savefig('top_companies.png')
plt.show()
```

**5. Relaci√≥n entre el tama√±o del grupo y el √©xito**

En esta 4 grafica, hemos podido observar la dispersi√≥n de relacci√≥n entre el n√∫mero de miembros originales y el estado de su actividad. </br>
Los grupos que m√°s miembros tienen suelen permanecer mas tiempo activos con lo cual suelen ser m√°s populares ya que est√°n m√°s tiempo producciendo musica.

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=all_groups_df, x='Members', y='Orig. Memb.', hue='Active', palette='muted')
plt.title('Tama√±o del grupo vs. √©xito', fontsize=14)
plt.xlabel('Current Members', fontsize=12)
plt.ylabel('Original Members', fontsize=12)
plt.legend(title='Active')
plt.savefig('group_size_vs_success.png')
plt.show()
```

**6. Predicci√≥n del √©xito basada en caracter√≠sticas pasadas**

EN esta grafica vemos los grupos que tienen un potencial de exito, la comparativa se hizo con grupos de 5 a 7 miembros. </br>
Con el tono oscuro observamos los grupos con mayor posibilidad de exito, dejando a DSP como el gupo mas exitoso. </br>
En el color m√°s claro los grupos con menor posibilidad de exito, estando en peor posicion el grupo Mafia.

```python
# Ejemplo de insight: los grupos m√°s exitosos suelen tener entre 5 y 7 miembros, pertenecer a ciertas compa√±√≠as o haber debutadoen a√±os espec√≠ficos.
successful_groups = all_groups_df[(all_groups_df['Active'] == 'Yes') & (all_groups_df['Members'] >= 5) & (all_groups_df['Members'] <= 7)]
predicted_groups = all_groups_df[(all_groups_df['Members'] >= 5) & (all_groups_df['Members'] <= 7) & (~all_groups_df['Name'].isin(successful_groups['Name']))]

plt.figure(figsize=(10, 6))
sns.barplot(data=predicted_groups, x='Members', y='Company', palette='rocket')
plt.title('Grupos potencialmente exitosos (5-7 miembros)', fontsize=14)
plt.xlabel('Number of Members', fontsize=12)
plt.ylabel('Company', fontsize=12)
plt.savefig('predicted_success_groups.png')
plt.show()
```

**7. Composici√≥n de g√©nero en la industria**

En esta gr√°fica se muestra una distribuci√≥n de genero en la industria del Kpop, ese√±ando una diferencia entre ambos generos. Siendo el genero masculino con mayor cantidad rozando casi los 700.

```python
gender_distribution = Kpop_idols['Gender'].value_counts().reset_index()
gender_distribution.columns = ['Gender', 'Count']

plt.figure(figsize=(8, 6))
sns.barplot(data=gender_distribution, x='Gender', y='Count', palette='pastel')
plt.title('Distribuci√≥n de g√©nero en la industria del K-POP', fontsize=14)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.savefig('gender_distribution.png')
plt.show()
```

**Grupos para hacer el festival de verano.**

```python
active_groups = all_groups_df[all_groups_df['Active'] == 'Yes']

top_5_groups = active_groups.sort_values(by='Members', ascending=False).head(5)[['Name', 'Company', 'Members']]

top_5_groups.reset_index(drop=True)
```

Para saber que grupos podr√≠an tener mas pospularidad en un festival de verano, escogimos los gupos musicales que mas mimbros tenian y los mas populares son actualmente.

**Grafica 5 Grupos para hacer el festival de verano.**

En esta ultima grafica, vemos que la poupularidad que tendria los grupos escogidos para el festival de verano estan bastante equiparados. Destacando auque por muy poco el grupo NCT con un 25.7%, el grupo que podria tener menos visitantes en el festival es el grupo The Boyz con un 17.1%.

```python
plt.figure(figsize=(8, 8))
plt.pie(
    top_5_groups['Members'],
    labels=top_5_groups['Name'],
    autopct='%1.1f%%',
    startangle=140,
    colors=sns.color_palette("pastel")
)
plt.title('TOP 5 grupos para un festival de verano', fontsize=14)
plt.savefig('top_5_groups_pie_chart.png')
plt.show()
```
